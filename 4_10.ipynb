```python
from statsmodels.tsa.arima_process import ArmaProcess
import matplotlib.pyplot as plt
import numpy as np
import yfinance as yf
import pandas as pd
```

```python
ar1 = ArmaProcess(ar=[1, -0.7]) # AR(1)
ma1 = ArmaProcess(ma=[1, 0.7]) # MA(1)
plt.figure(figsize=(10,4))
plt.plot(ar1.generate_sample(nsample=200), label='AR(1)', alpha=0.7)
plt.plot(ma1.generate_sample(nsample=200), label='MA(1)', alpha=0.7)
plt.legend()
```
